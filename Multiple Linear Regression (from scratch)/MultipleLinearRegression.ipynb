```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
v1Happiness = pd.read_csv("https://raw.githubusercontent.com/lauradiosan/AI-UBB/main/2022-2023/labs/lab06/data/v1_world-happiness-report-2017.csv")
v2Happiness = pd.read_csv("https://raw.githubusercontent.com/lauradiosan/AI-UBB/main/2022-2023/labs/lab06/data/v2_world-happiness-report-2017.csv")
v3Happiness = pd.read_csv("https://raw.githubusercontent.com/lauradiosan/AI-UBB/main/2022-2023/labs/lab06/data/v3_world-happiness-report-2017.csv")
display(v1Happiness)
```

```python

plt.scatter(v1Happiness["Economy..GDP.per.Capita."], v1Happiness["Happiness.Score"])
plt.ylabel("Happiness")
plt.xlabel("GDP")
plt.show()

plt.scatter(v1Happiness["Freedom"], v1Happiness["Happiness.Score"])
plt.ylabel("Happiness")
plt.xlabel("Freedom")
plt.show()
```

```python
#prepare_data
X = v1Happiness[["Economy..GDP.per.Capita.", "Freedom"]].values #my features 'matrix'
y = v1Happiness[["Happiness.Score"]].values #my output
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.2, random_state=0)
sklearn_model = LinearRegression()
sklearn_model.fit(X_train, y_train)

y_computed = sklearn_model.predict(X_validation)
```

```python
indexes = [i for i in range(len(y_validation))]
real, = plt.plot(indexes, y_validation, 'ro', label = 'real')
computed, = plt.plot(indexes, y_computed, 'bo', label = 'computed')
plt.legend([real, (real, computed)], ["Real", "Computed"])
plt.show()
```
